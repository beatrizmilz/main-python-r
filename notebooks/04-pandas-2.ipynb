```python
# Obter os files com gênero 'Drama' e ordernar pela quantidade de
# vezes que a palavra 'woman' (mulher) aparece na descrição.
imdb[imdb.generos.str.contains("Drama")].assign(
    qnt_woman = imdb.descricao.str.lower().str.count("woman")
).sort_values("qnt_woman", ascending=False).iloc[0].descricao
```

### Summarise

```python
imdb.filter(["duracao", "orcamento", "receita"]).agg(["mean", "sum", "count"])
```

```python
imdb.agg({
  "duracao": ["mean"],
  "orcamento": ["mean", "count"]
})
```

```python
imdb.groupby(["ano"], as_index = False).agg(
  duracao_media = ("duracao", "mean"),
  desv_pad_duracao = ("duracao", "std"),
  receita_max = ("receita", max)
)
```

```python
imdb.groupby('ano', as_index=False).apply(lambda x: pd.DataFrame({
    "duracao_media": [x.duracao.mean()],
    "desv_pad_duracao": [x.duracao.std()],
    "receita_max": [x.receita.max()],
    "lucro_medio": [(x.receita - x.orcamento).mean()]
}))
```

### Exercícios

```python
# Obtenha os 10 anos com as as maiores notas médias do IMDB.
# Coloque em um data frame ordeno pela média das notas.
imdb['nota_imdb_numeric'] = imdb['nota_imdb'].str.replace(",", ".").astype(float)


imdb.groupby("ano", as_index=False).agg(nota_media = ("nota_imdb_numeric", "mean")).sort_values("nota_media", ascending=False).iloc[:9]
```

```python
# Obtenha a proporção de filmes que deram prejuizo por ano.

# OBS: CONSIDEREI SÓ FILMES QUE TEMOS AS INFORMAÇÕES DE LUCRO E PREJUIZO!!

quantidade_filmes_prejuizo = imdb.assign(
    prejuizo = imdb.receita < imdb.orcamento
).groupby(["ano", "prejuizo"], as_index=False).agg(
    quantidade_prejuizo = ("prejuizo", "count")
).query("prejuizo == True").filter(["ano", "quantidade_prejuizo"])


quantidade_filmes = imdb.dropna(subset=["orcamento", 'receita'], axis=0).groupby(["ano"], as_index=False).agg(
    quantidade_filme = ("id_filme", "count")
)

# ter uma coluna 

quantidade_filmes_prejuizo
quantidade_filmes

imdb_com_ano_preju = pd.merge(quantidade_filmes_prejuizo, quantidade_filmes, how = "left", on = ["ano"])

resposta_imdb_preju = imdb_com_ano_preju.assign(
    proporcao = (imdb_com_ano_preju.quantidade_prejuizo / imdb_com_ano_preju.quantidade_filme) * 100
)


resposta_imdb_preju
```

### Left join

```python
lucro_direcao = (imdb.
    assign(
        lucro = lambda x: x.receita - x.orcamento
    ).
    groupby(["direcao"], as_index = False).
    agg(
        lucro_medio = ('lucro', 'mean')
    )
)
lucro_direcao
```

```python
imdb_com_lucro = pd.merge(imdb, lucro_direcao, how = "left", on = ["direcao"])

imdb_com_lucro
```

```python
(imdb_com_lucro.
    assign(lucro = lambda x: x.receita - x.orcamento).
    filter(["titulo", "lucro", "lucro_medio"]).
    query("lucro > lucro_medio")
)
```
